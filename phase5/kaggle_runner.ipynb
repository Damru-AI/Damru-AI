# 🐕 Damru Learning Engine — Kaggle Free Runner

Ek aur **FREE 24/7-style worker**. Kaggle har session ~12 ghante chalata hai.

### Setup (sirf ek baar)
1. Kaggle > Notebook > **Add-ons > Secrets** me ye daalo:
   `SUPABASE_URL`, `SUPABASE_KEY`, `OPENROUTER_KEY`, `GROQ_API_KEY`, `GEMINI_KEY` (jo hain).
2. Settings me **Internet = ON** karo.
3. Upar **Save Version > Save & Run All (Commit)** — ye background me 12h tak chalega.

> Ye GitHub Actions ke alawa EXTRA data banata hai. SHARD_ID=0/SHARD_TOTAL=1 = full coverage.


```python
# 1) Get the engine code
!rm -rf Damru-AI && git clone --depth 1 https://github.com/Damru-AI/Damru-AI.git
%cd Damru-AI
!pip -q install -r phase5/requirements.txt
```

```python
# 2) Load secrets from Kaggle (Add-ons > Secrets)
import os
try:
    from kaggle_secrets import UserSecretsClient
    us = UserSecretsClient()
    for k in ['SUPABASE_URL','SUPABASE_KEY','OPENROUTER_KEY','GROQ_API_KEY','GEMINI_KEY']:
        try:
            os.environ[k] = us.get_secret(k)
        except Exception as e:
            print('missing secret (ok if unused):', k)
    os.environ['OPENROUTER_API_KEY'] = os.environ.get('OPENROUTER_KEY','')
    print('secrets loaded')
except Exception as e:
    print('Not on Kaggle or no secrets:', e)
```

```python
# 3) Run the engine (stops itself a little before Kaggle's 12h limit)
os.environ['SHARD_ID'] = '0'
os.environ['SHARD_TOTAL'] = '1'
os.environ['MAX_RUNTIME_MIN'] = '690'   # ~11.5h
os.environ['GENERAL_WORKERS'] = '8'
os.environ['CODING_WORKERS'] = '12'
!python phase5/run_engine.py
```
